```python
#6-b: Write a program to demonstrate use of groupby() method.

# importing pandas module
import pandas as pd

# Define a dictionary containing employee data
data1 = {'Name':['Srinidhi', 'Anu', 'Jagath', 'Sharmila','Bhavana', 'Aditya', 'Mahesh',    'Prabhas'],
		'Age':[27, 24, 22, 32,33, 36, 27, 32],
		'Address':['Nagpur', 'Kanpur', 'Allahabad', 'Tekkali','Srikakulam', 'Kanpur', 'Allahabad', 'Palasa'],
		'Qualification':['Msc', 'MA', 'MCA', 'Phd','B.Tech', 'B.com', 'Msc', 'MA']}
	
# Convert the dictionary into DataFrame
df = pd.DataFrame(data1)
print(df)

# using groupby function with one key
df.groupby('Name')
print(df.groupby('Name').groups)

# Using multiple keys in groupby() function
df.groupby(['Name', 'Qualification'])
print(df.groupby(['Name', 'Qualification']).groups)
```

```python
# 6-c: Write a program to demonstrate pandas Merging, Joining and Concatenating

# importing pandas module
import pandas as pd 
 
# Define a dictionary containing employee data 
data1 = {'Name':['Gangotri', 'Karthik', 'Naresh', 'Pravallika'], 
        'Age':[27, 24, 22, 32], 
        'Address':['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'], 
        'Qualification':['Msc', 'MA', 'MCA', 'Phd']} 
   
# Define a dictionary containing employee data 
data2 = {'Name':['Manoj', 'Varma', 'Kishore', 'Ritesh'], 
        'Age':[17, 14, 12, 52], 
        'Address':['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'], 
        'Qualification':['Btech', 'B.A', 'Bcom', 'B.hons']} 
 
# Converting dictionaries to dataframes
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
#df3 = pd.DataFrame(data3)

# Merging
merged_df = pd.merge(df1, df2, on='Address')
print("Merged DataFrame:")
print(merged_df)

# Joining
df1.set_index('Address', inplace=True)
df2.set_index('Address', inplace=True)
joined_df = df1.join(df2, lsuffix='_left', rsuffix='_right')
print("\nJoined DataFrame:")
print(joined_df)

# Concatenating
concatenated_df = pd.concat([df1, df2], axis=1)
print("\nConcatenated DataFrame:")
print(concatenated_df)
```

```python
#6-a:Write a program which use Pandas inbuilt visualization to plot following graphs: i) Bar plots ii) Histograms  iii) Line plots  iv) Scatter plots

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Sample data
data = {
    'Category': ['A', 'B', 'C', 'D', 'E'],
    'Value1': [10, 20, 15, 25, 30],
    'Value2': [5, 15, 10, 20, 25]
}

# Create a DataFrame
df = pd.DataFrame(data)

# i) Bar plots
df.plot(kind='bar', x='Category', y=['Value1', 'Value2'], title='Bar Plot')
plt.show()

# ii) Histograms
df['Value1'].plot(kind='hist', bins=10, title='Histogram')
plt.show()

# iii) Line plots
df.plot(kind='line', x='Category', y=['Value1', 'Value2'], marker='o', title='Line Plot')
plt.show()

# iv) Scatter plots
df.plot(kind='scatter', x='Value1', y='Value2', title='Scatter Plot')
plt.show()
```
